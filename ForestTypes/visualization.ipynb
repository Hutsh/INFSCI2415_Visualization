```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import data

```python
train = pd.read_csv('training.csv')
test = pd.read_csv('testing.csv')
data = train.append(test,ignore_index=True)
```

## Distribution

```python
fig, ax = plt.subplots(dpi=100)
sns.set(style="darkgrid")
ax = sns.countplot(x="class", data=data)
ax.set_xticklabels(\
                   ['d(Mixed deciduous\' forest)','h(Hinoki forest)','s(Sugi forest)','o(Other non-forest land)']\
                   ,rotation=15)
```

```python
sep_2010 = data[['class','b1','b2','b3']]
mar_2011 = data[['class','b4','b5','b6']]
may_2011 = data[['class','b7','b8','b9']]
```

# X:class, Y value

```python
fig, ax = plt.subplots(dpi=100)
sns.scatterplot(sep_2010['class'], sep_2010['b1'], s=200, edgecolor='k',facecolor='g', alpha=0.5, ax=ax)
sns.scatterplot(sep_2010['class'], sep_2010['b2'], s=200, edgecolor='k',facecolor='r', alpha=0.5, ax=ax)
sns.scatterplot(sep_2010['class'], sep_2010['b3'], s=200, edgecolor='k',facecolor='b', alpha=0.5, ax=ax)
```

```python
# sep_2010 = data[['class','b1','b2','b3']]
# mar_2011 = data[['class','b4','b5','b6']]
# may_2011 = data[['class','b7','b8','b9']]
sep_2010_melt = pd.melt(sep_2010, var_name='band', value_name='value',id_vars=['class'])
class_dic = {'d ': 'Mixed deciduous forest', 'h ': 'Hinoki forest', 's ': 'Sugi forest','o ': 'Other non-forest land'}
band_dic = {'b1' : 'green', 'b2': 'red', 'b3': 'near infrared'}
sep_2010_melt.replace(class_dic, inplace=True)
sep_2010_melt.replace(band_dic, inplace=True)
fig, ax = plt.subplots(dpi=100)
plt.xticks(rotation=15)
g = sns.stripplot(data = sep_2010_melt, x='class', y = 'value', hue = 'band', 
              jitter = 0.05, dodge = True, alpha = 0.8, 
              palette = ['g','r','k'])
ax.legend(bbox_to_anchor=(1, 1))
ax.set_title('band value of different forest class (Sept. 26, 2010)')
```
